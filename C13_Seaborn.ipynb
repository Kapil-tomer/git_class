```python
#co relation matrix:: 
#heat map 


#cormat =pokemon.corr()
#cormat

# Drop non-numeric columns
pokemon_numeric = pokemon.select_dtypes(include=[np.number])

# Calculate correlation matrix
cormat = pokemon_numeric.corr()
print(cormat)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot heatmap
sns.heatmap(cormat,vmax=0.8, annot=True, fmt= '0.2f',cmap='viridis')

# Display the heatmap
plt.show()
```
