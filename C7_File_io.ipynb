```python
#OPEN, READ AND CLOSE FILE
#OPEN: BEFORE READING AND WRITING WE HAVE TO OPEN A FILE
#open("file_name","mode") , Mode:: r:read, w:write, by default: read
f=open("C3_LISTS.ipynb","r")
data=f.read()
print(data)
f.close()
```

```python
# 'r' : OPEN FOR READING , DEFAULT
# 'w' : OPEN FOR WRITING, TURNICATING(DELETING THE PREVIOUS FILE')
# 'x' : CRAETE A NEW FILE AND OPEN FOR WRITING
# 'a' : append, OPEN FOR WRITING, APPENDING AT THE END OF THE FILE IF IT EXISTS
# 'b' : BINARY MODE
# 't' : TEXT MODE (default)
# "+' : OPEN A DISK FILE FOR (READ AND WRITE), EG: r+
```
