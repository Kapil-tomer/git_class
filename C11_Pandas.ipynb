```python
#merging ::::::

import pandas as pd

file1=pd.DataFrame({"a":[1,2,3,4],"b":[9,9,9,9]})
file2=pd.DataFrame({"a":[3,6,5,5],"b":[4,8,3,5]})


print(pd.merge(file1,file2))
print(pd.merge(file1,file2,on="a"))
print(pd.merge(file1,file2,how="left"))
print(pd.merge(file1,file2,how="right"))     
print(pd.merge(file1,file2,how="outer") )      

print(pd.merge(file1,file2,suffixes=("name","pyhton")))






```

```python
#MELT
var1=pd.DataFrame({"days":[1,2,3,4],"eng":[5,6,7,8],"math":[16,17,18,20]})
var1
```

```python
#pivot::

var1=pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","b","c"],"eng":[11,13,14,15,16,17],"maths":[16,17,18,19,15,14]})
var1
```
